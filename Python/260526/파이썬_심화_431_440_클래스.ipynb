import time

```python
import time
class Cache:
  def __init__(self,ttl=60):
    self._data={}
    self._ttl=ttl
  def set(self,key,value):
    self._data[key]=(value,time.time()+self._ttl)
  def get(self,key):
    if key in self._data:
      val,exp=self._data[key]
      if time.time()<exp:
        return val
      del self._data
    return None
  def __repr__(self):
    return f'Cache({len(self._data)} items)'
c=Cache(ttl=2)
c.set('user_1','김파이')
print(c.get('user_1'))   # 김파이
time.sleep(1)
print(c.get('user_1'))   # 김파이 (TTL 유지)
time.sleep(2)
print(c.get('user_1'))   # None (TTL 만료)
```

### 434 클래스 - 커맨드 패턴

실행 취소(undo)가 가능한 커맨드 패턴을 구현하세요.

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
class TextEditor:
    def __init__(self): self.text=''; self.history=[]
    def execute(self,command): command.execute(self); self.history.append(command)
    def undo(self):
        if self.history: self.history.pop().undo(self)

class AddText:
    def __init__(self,text): self.text=text
    def execute(self,editor): editor.text+=self.text
    def undo(self,editor): editor.text=editor.text[:-len(self.text)]

editor=TextEditor()

editor.execute(AddText('하세요'))
print(editor.text)  # 안녕하세요

</pre>
</details>

```python
class TextEditor:
  def __init__(self):
    self.text=''
    self.history=[]
  def execute(self,command):
    command.execute(self)
    self.history.append(command)
  def undo(self):
    if self.history:
      self.history.pop().undo(self)
class AddText:
  def __init__(self,text):
    self.text=text
  def execute(self,editor):
    editor.text+=self.text
  def undo(self,editor):
    editor.text=editor.text[:-len(self.text)]
editor=TextEditor()
editor.execute(AddText('안녕'))
editor.execute(AddText('하세요'))
print(editor.text)  # 안녕하세요
editor.undo()
print(editor.text)  # 안녕
editor.undo()
print(editor.text)  # (빈 문자열)
```

```python
class State:
  def handle(self,traffic_light):
    pass

class RedState(State):
  def handle(self,tl):
    print('🔴 정지')
    tl.state=GreenState()


class GreenState(State):
  def handle(self,tl):
    print('🟢 통행')
    tl.state=YellowState()


class YellowState(State):
  def handle(self,tl):
    print('🟡 주의')
    tl.state=RedState()


class TrafficLight:
  def __init__(self):
    self.state=RedState()
  def change(self):
    self.state.handle(self)


tl=TrafficLight()
for _ in range(6):
  tl.change()
```

```python
import math

class GameObject:
  def __init__(self,name,x=0,y=0):
    self.name=name
    self.x=x; self.y=y
  def move(self,dx,dy):
    self.x+=dx
    self.y+=dy
  def distance_to(self,other):
    return math.sqrt((self.x-other.x)**2+(self.y-other.y)**2)
  def repr(self):
    return f'{self.name}({self.x},{self.y})'


class Player(GameObject):
  def __init__(self,name):
    super().__init__(name)
    self.hp=100
  def attack(self,target,dmg):
    print(f'{self.name} → {target.name} 공격! {dmg}데미지')
    target.hp=max(0,target.hp-dmg)


p=Player('영웅')
e=Player('몬스터')
e.move(3,4)
print(f'거리: {p.distance_to(e):.1f}')
p.attack(e,30)
print(f'{e.name} HP: {e.hp}')
```

```python
class RealDatabase:
  def query(self,sql):
    print(f'DB 쿼리 실행: {sql}')
    return f'결과({sql})'

class DatabaseProxy:
  def __init__(self,user_level):
    self._db=RealDatabase()
    self._level=user_level
  def query(self,sql):
    if 'DROP' in sql.upper() and self._level<3:
      print('권한 없음: DROP 명령어는 관리자만 실행 가능')
      return None
    return self._db.query(sql)


user_db=DatabaseProxy(user_level=1)
admin_db=DatabaseProxy(user_level=3)
user_db.query('SELECT * FROM users')
user_db.query('DROP TABLE users')
admin_db.query('DROP TABLE temp_data')
```

class BubbleSort:
    def sort(self,data):
        d=data[:]
        for i in range(len(d)):

```python
class BubbleSort:
  def sort(self,data):
    d=data[:]
    for i in range(len(d)):
      for j in range(len(d)-i-1):
        if d[j]>d[j+1]: d[j],d[j+1]=d[j+1],d[j]
    return d

class PythonSort:
  def sort(self,data):
    return sorted(data)

class Sorter:
  def __init__(self,strategy):
    self.strategy=strategy
  def sort(self,data):
    return self.strategy.sort(data)

data=[3,1,4,1,5,9,2,6]
for strategy in [BubbleSort(),PythonSort()]:
  s=Sorter(strategy)
  print(type(strategy).__name__,':', s.sort(data))
```

```python
class Router:
  def __init__(self): self._routes={}
  def route(self,path,method='GET'):
    def decorator(func):
      self._routes[(method,path)]=func
      return func
    return decorator
  def dispatch(self,method,path,**kwargs):
    handler=self._routes.get((method,path))
    if handler:
      return handler(**kwargs)
    return '404 Not Found'

app=Router()


@app.route('/')
def index():
  return '''

홈페이지
'''

@app.route('/user',method='GET')
def get_user(name='게스트'): return f'사용자: {name}'

@app.route('/user',method='POST')
def create_user(name=''): return f'{name} 생성 완료'

print(app.dispatch('GET','/'))
print(app.dispatch('GET','/user',name='김파이'))
print(app.dispatch('POST','/user',name='이파이'))
print(app.dispatch('GET','/unknown'))
```
