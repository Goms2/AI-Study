### 333 딕셔너리로 환율 변환기

딕셔너리 환율을 활용해 금액을 원화로 변환하는 함수를 만드세요.
```python
rates={'USD':1300,'EUR':1400,'JPY':9.0,'CNY':180}
# convert(100,'USD') → 130000원
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
rates={'USD':1300,'EUR':1400,'JPY':9.0,'CNY':180}
def convert(amount,currency):
    if currency not in rates:
        return '지원하지 않는 통화'
    return amount * rates[currency]
print(convert(100,'USD'))  # 130000
print(convert(1000,'JPY')) # 9000.0
</pre>
</details>

````python
rates={'USD':1300,'EUR':1400,'JPY':9.0,'CNY':180}
def convert(amount, currency):
  if currency not in rates:
    return print('없음')
  if currency in rates:
    return amount * rates[currency]
print(f"{convert(100,'USD'):,}원")
print(f"{int(convert(10000,'JPY')):,}원")
````

### 335 딕셔너리로 성적 등급 처리

점수 딕셔너리를 성적 등급 딕셔너리로 변환하세요.
```python
scores={'민수':85,'지영':62,'철수':91,'수연':55,'태호':78}
# 90이상:A, 80이상:B, 70이상:C, 60이상:D, 60미만:F
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
scores={'민수':85,'지영':62,'철수':91,'수연':55,'태호':78}
def grade(s):
    if s>=90: return 'A'
    elif s>=80: return 'B'
    elif s>=70: return 'C'
    elif s>=60: return 'D'
    else: return 'F'
grades={name:grade(s) for name,s in scores.items()}
print(grades)
</pre>
</details>

````python
scores={'민수':85,'지영':62,'철수':91,'수연':55,'태호':78}
def 등급(v):
  if v >= 90:
    return 'A'
  elif v >= 80:
    return 'B'
  elif v >= 70:
    return 'C'
  elif v >= 60:
    return 'D'
  else:
    return 'F'
grades={name:등급(v) for name,v in scores.items()}
print(grades)
````

### 337 딕셔너리로 스택 구현

딕셔너리를 사용해 이름이 붙은 스택을 구현하세요. push, pop, peek 기능을 구현하세요.

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
class NamedStack:
    def __init__(self, name):
        self.name = name
        self.data = []
    def push(self, val): self.data.append(val)
    def pop(self): return self.data.pop() if self.data else None
    def peek(self): return self.data[-1] if self.data else None

stacks={'A':NamedStack('A'),'B':NamedStack('B')}
stacks['A'].push(1); stacks['A'].push(2)
stacks['B'].push(10)
print(stacks['A'].pop())  # 2
print(stacks['B'].peek()) # 10
</pre>
</details>

````python
class NamedStack:
  def __init__(self, name):
    self.name = name
    self.data = []
  def push(self, x):
    self.data.append(x)
  def pop(self):
    return self.data.pop() if self.data else None
  def peek(self):
    return self.data[-1] if self.data else None

stacks={'A':NamedStack('A'),'B':NamedStack('B')}
stacks['A'].push(1); stacks['A'].push(2)
stacks['B'].push(10)
print(stacks['A'].pop())  # 2
print(stacks['B'].peek()) # 10
````

### 339 딕셔너리로 투표 집계

투표 결과 리스트에서 후보별 득표수와 승자를 출력하세요.
```python
votes=['김파이','이파이','김파이','박파이','이파이','김파이','김파이']
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
from collections import Counter
votes=['김파이','이파이','김파이','박파이','이파이','김파이','김파이']
result=Counter(votes)
for name,cnt in result.most_common():
    print(f'{name}: {cnt}표')
print('당선:', result.most_common(1)[0][0])
</pre>
</details>

````python
from collections import Counter
votes=['김파이','이파이','김파이','박파이','이파이','김파이','김파이']
result=Counter(votes)
for name, n in result.most_common():
  print(f'{name}: {n}표')
print('승자:', result.most_common(1)[0][0])
````

### 340 딕셔너리 직렬화 (JSON)

````python
import json
data={'name':'김파이','scores':[85,92,78],'active':True}
JSON=json.dumps(data, ensure_ascii=False)
print(type(JSON), JSON)

back=json.loads(JSON)
print(type(back), back)
````
