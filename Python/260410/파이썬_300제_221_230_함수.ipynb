### 221 문자열 역순 출력 함수

입력된 문자열을 역순으로 출력하는 print_reverse 함수를 정의하라.
```
실행 예:
print_reverse("python")

```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>

<pre>

print_reverse("python")
</pre>
</details>

````python
def print_reverse(문자열):
  print(문자열[::-1])
print_reverse("python")
````

성적 리스트를 입력받아 평균을 출력하는 print_score 함수를 정의하라.
```
실행 예:
```

<details>

### 223 짝수만 출력 함수

하나의 리스트를 입력받아 짝수만 화면에 출력하는 print_even 함수를 정의하라.
```
실행 예:
print_even([1, 3, 2, 10, 12, 11, 15])
2
10
12
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
for문과 if문을 함께 사용합니다.</br>
<pre>
def print_even(my_list):
    for v in my_list:
        if v % 2 == 0:
            print(v)

print_even([1, 3, 2, 10, 12, 11, 15])
</pre>
</details>

````python
def print_even(n):
  for i in n:
    if i % 2 == 0:
      print(i)
print_even([1, 3, 2, 10, 12, 11, 15])
````

### 224 딕셔너리 키 출력 함수

하나의 딕셔너리를 입력받아 딕셔너리의 key 값을 화면에 출력하는 print_keys 함수를 정의하라.
```
실행 예:
print_keys({"이름":"김말똥", "나이":30, "성별":0})
이름
나이
성별
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
<pre>
def print_keys(dic):
    for key in dic.keys():
        print(key)

print_keys({"이름":"김말똥", "나이":30, "성별":0})
</pre>
</details>

````python
def print_keys(문자열):
  for i in 문자열.keys():
    print(i)

print_keys({"이름":"김말똥", "나이":30, "성별":0})
````

### 226 5글자씩 출력 함수

입력 문자열을 한 줄에 다섯 글자씩 출력하는 print_5xn(string) 함수를 작성하라.
```
실행 예:
print_5xn("아이엠어보이유알어걸")
아이엠어보
이유알어걸
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
슬라이싱으로 5글자씩 잘라서 출력합니다.</br>
<pre>
def print_5xn(line):
    chunk_num = int(len(line) / 5)
    for x in range(chunk_num + 1):
        print(line[x * 5: x * 5 + 5])

print_5xn("아이엠어보이유알어걸")
</pre>
</details>

````python
def print_5xn(문자열):
  for i in range(0,len(문자열),5):
    print(문자열[i:i+5])
print_5xn("아이엠어보이유알어걸")
````

### 227 n글자씩 출력 함수

문자열과 한 줄에 출력될 글자 수를 입력받아 한 줄에 해당 글자 수만큼 출력하는 print_mxn(string, num) 함수를 작성하라.
```
실행 예:
print_mxn("아이엠어보이유알어걸", 3)
아이엠
어보이
유알어
걸
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
226번을 일반화한 함수입니다.</br>
<pre>
def print_mxn(line, num):
    chunk_num = int(len(line) / num)
    for x in range(chunk_num + 1):
        print(line[x * num: x * num + num])

print_mxn("아이엠어보이유알어걸", 3)
</pre>
</details>

````python
def print_mxn(문자열, num):
  for i in range(0,len(문자열),num):
    print(문자열[i:i+num])
print_mxn("아이엠어보이유알어걸", 3)
````

### 228 월급 계산 함수 (return)

연봉을 입력받아 월급을 계산하는 calc_monthly_salary(annual_salary) 함수를 정의하라. 연봉을 12개월로 나누며 1원 미만은 버림한다.
```
실행 예:
calc_monthly_salary(12000000)
1000000
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
return으로 계산 결과를 반환합니다. int()로 소수점 이하를 버림합니다.</br>
<pre>
def calc_monthly_salary(annual_pay):
    monthly_pay = int(annual_pay / 12)
    return monthly_pay

result = calc_monthly_salary(12000000)
print(result)
</pre>
</details>

````python
def calc_monthly_salary(연봉):
  print(int(연봉/12))
calc_monthly_salary(12000000)
````

### 229 키워드 인자 - 결과 예측

아래 코드의 실행 결과를 예측하라.
```python
def my_print(a, b):
    print("왼쪽:", a)
    print("오른쪽:", b)

my_print(a=100, b=200)
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
키워드 인자(keyword argument)는 파라미터 이름을 명시해서 전달합니다. 순서와 상관없이 이름으로 바인딩됩니다.</br>
<pre>
# 결과:
# 왼쪽: 100
# 오른쪽: 200
</pre>
</details>

````python
def my_print(a, b):
    print("왼쪽:", a)
    print("오른쪽:", b)

my_print(a=100, b=200)
````
