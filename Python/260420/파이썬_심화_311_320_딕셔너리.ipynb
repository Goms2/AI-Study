### 314 학점별 그룹화

학생 리스트를 학점별로 그룹화하세요.
```python
students=[('민수','A'),('지영','B'),('철수','A'),('수연','C'),('태호','B')]
# 결과: {'A':['민수','철수'],'B':['지영','태호'],'C':['수연']}
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
students=[('민수','A'),('지영','B'),('철수','A'),('수연','C'),('태호','B')]

for name,grade in students:
    groups.setdefault(grade,[]).append(name)
print(groups)
</pre>
</details>

````python
students=[('민수','A'),('지영','B'),('철수','A'),('수연','C'),('태호','B')]
groups={}
for name,grade in students:
    groups.setdefault(grade,[]).append(name)
print(groups)
````

### 315 최댓값/최솟값 키 찾기

딕셔너리에서 가장 높은 점수와 가장 낮은 점수의 학생을 출력하세요.
```python
scores={'민수':85,'지영':62,'철수':91,'수연':55,'태호':78}
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
scores={'민수':85,'지영':62,'철수':91,'수연':55,'태호':78}
print('최고:', max(scores, key=scores.get))
print('최저:', min(scores, key=scores.get))
</pre>
</details>

````python
scores={'민수':85,'지영':62,'철수':91,'수연':55,'태호':78}
top=max(scores, key=scores.get)
bottom=min(scores, key=scores.get)
print(f"가장 높은 점수의 학생: {top}, 가장 낮은 점수의 학생: {bottom}")
````

### 316 딕셔너리 깊은 복사

얕은 복사와 깊은 복사의 차이를 확인하세요.
```python
original = {'name':'김파이', 'scores':[85,90,78]}
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
import copy
d={'name':'김파이','scores':[85,90,78]}

### 317 딕셔너리로 통계 계산

주식 데이터 리스트에서 종목별 평균 종가를 계산하세요.
```python
data=[{'code':'005930','close':70000},{'code':'005930','close':71000},
      {'code':'035420','close':350000},{'code':'035420','close':360000}]
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
data=[{'code':'005930','close':70000},{'code':'005930','close':71000},
      {'code':'035420','close':350000},{'code':'035420','close':360000}]
totals={}
counts={}
for item in data:
    c=item['code']
    totals[c]=totals.get(c,0)+item['close']
    counts[c]=counts.get(c,0)+1
for c in totals:
    print(c, f'{totals[c]/counts[c]:,.0f}원')
</pre>
</details>

````python
data=[{'code':'005930','close':70000},{'code':'005930','close':71000},
      {'code':'035420','close':350000},{'code':'035420','close':360000}]
totals={}
counts={}
for item in data:
    c=item['code']
    totals[c]=totals.get(c,0)+item['close']
    counts[c]=counts.get(c,0)+1
for c in totals:
  print(c, f"{totals[c]/counts[c]:,.0f}원")
````

### 318 중첩 딕셔너리에 평균 추가

학생별 과목 점수 딕셔너리에 'avg' 키를 추가하세요.
```python
students={'민수':{'국어':85,'수학':92},'지영':{'국어':90,'수학':65}}
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
students={'민수':{'국어':85,'수학':92},'지영':{'국어':90,'수학':65}}
for name,scores in students.items():
    scores['avg']=round(sum(scores.values())/len(scores),1)
print(students)
</pre>
</details>

````python
students={'민수':{'국어':85,'수학':92},'지영':{'국어':90,'수학':65}}
for name, scores in students.items():
    scores['avg']=round(sum(scores.values())/len(scores),1)
print(students)
````

### 319 defaultdict 활용

collections.defaultdict를 사용해 단어 빈도수를 구하세요.
```python
words=['apple','banana','apple','cherry','banana','apple']
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
from collections import defaultdict
words=['apple','banana','apple','cherry','banana','apple']
count=defaultdict(int)
for w in words:
    count[w]+=1
print(dict(count))
</pre>
</details>

````python
from collections import defaultdict
words=['apple','banana','apple','cherry','banana','apple']
count=defaultdict(int)
for w in words:
    count[w]+=1
print(dict(count))
````

### 320 딕셔너리 역순 출력

딕셔너리를 삽입 역순으로 출력하세요.
```python
d = {'c':3,'a':1,'b':2}
# 역순: b,a,c
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
d={'c':3,'a':1,'b':2}
for k in reversed(list(d.keys())):
    print(k, d[k])
</pre>
</details>

````python
d = {'c':3,'a':1,'b':2}
t={}
for k in reversed(list(d.keys())):
    d1={k:d[k]}
    t.update(d1)
print(t)
````
