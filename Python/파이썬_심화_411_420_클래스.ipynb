### 412 팩토리 메서드 패턴

클래스 메서드를 팩토리로 사용해 다양한 방식으로 객체를 생성하세요.
```python
# 문자열로 생성: Person.from_string('김파이,25')
# 딕셔너리로 생성: Person.from_dict({'name':'김파이','age':25})
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
class Person:
    def __init__(self,name,age): self.name=name; self.age=age
    @classmethod
    def from_string(cls,s):
        name,age=s.split(',')
        return cls(name,int(age))
    @classmethod
    def from_dict(cls,d):
        return cls(d['name'],d['age'])
    def __repr__(self): return f'Person({self.name},{self.age})'

p1=Person.from_string('김파이,25')
p2=Person.from_dict({'name':'이파이','age':30})

</details>

````python
class Person:
  def __init__(self,name,age):
    self.name=name
    self.age=age
  @classmethod
  def from_string(cls,s):
    name,age=s.split(',')
    return cls(name,int(age))
  @classmethod
  def from_dict(cls,d):
    return cls(d['name'], d['age'])
  def __repr__(self):
    return f'{self.name}, {self.age}'
print(Person.from_string('김파이,25'))
print(Person.from_dict({'name':'김파이','age':25}))
````

### 413 이터레이터 클래스

__iter__와 __next__를 구현해서 직접 이터레이터를 만드세요.
```python
# 1부터 n까지 숫자를 하나씩 반환하는 이터레이터
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
class Counter:
    def __init__(self,n): self.n=n; self.current=1
    def __iter__(self): return self
    def __next__(self):
        if self.current>self.n: raise StopIteration
        val=self.current; self.current+=1
        return val

for i in Counter(5): print(i, end=' ')
print()
print(list(Counter(5)))
</pre>
</details>

````python
class Counter:
  def __init__(self,n):
    self.n=n
    self.current=1
  def __iter__(self):
    return self
  def __next__(self):
    if self.current>self.n:
      raise StopIteration
    val=self.current
    self.current+=1
    return val
for i in Counter(10):
  print(i, end=' ')
print()
print(list(Counter(10)))
````

class FibGenerator:
    def __init__(self,n): self.n=n
    def __iter__(self):
        a,b=0,1

````python
class FibGenerator:
  def __init__(self, n):
    self.n=n
  def __iter__(self):
    a,b=0,1
    for _ in range(self.n):
      yield a
      a,b=b,a+b
for f in FibGenerator(10):
    print(f, end=' ')
print()
````

import time

### 416 데이터클래스 (dataclass)

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
from dataclasses import dataclass, field

@dataclass
class Student:
    name: str
    age: int
    scores: list = field(default_factory=list)

    def avg(self):
        return sum(self.scores)/len(self.scores) if self.scores else 0

````python
from dataclasses import dataclass, field

@dataclass
class Student:
  name: str
  age: int
  scores: list = field(default_factory=list)


  def avg(self):
    return sum(self.scores)/len(self.scores) if self.scores else 0
s1=Student('김파이',20,[85,90,78])
s2=Student('이파이',22,[92,88,95])
print(s1)
print(f'{s1.name} 평균: {s1.avg():.1f}')
print(s1==s2)
````

### 417 Mixin 패턴

Mixin 클래스를 사용해서 직렬화 기능을 여러 클래스에 추가하세요.

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
import json

class JSONMixin:
    def to_json(self):
        return json.dumps(self.__dict__, ensure_ascii=False)
    @classmethod
    def from_json(cls,json_str):
        return cls(**json.loads(json_str))

class Person(JSONMixin):
    def __init__(self,name,age):
        self.name=name; self.age=age

class Product(JSONMixin):
    def __init__(self,name,price):
        self.name=name; self.price=price

p=Person('김파이',25)
print(p.to_json())

product=Product('사과',3000)
print(product.to_json())
</pre>
</details>

````python
import json

class JSONMixin:
  def to_json(self):
    return json.dumps(self.__dict__, ensure_ascii=False)
  @classmethod
  def from_json(cls,json_str):
    return cls(**json.loads(json_str))

class Person(JSONMixin):
  def __init__(self,name,age):
    self.name=name
    self.age=age

class Product(JSONMixin):
  def __init__(self,name,price):
    self.name=name
    self.price=price

p=Person('김파이',25)
print(p.to_json())

product=Product('사과',3000)
print(product.to_json())
````

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
class Node:
    def __init__(self,val): self.val=val; self.next=None

class LinkedList:
    def __init__(self): self.head=None
    def append(self,val):
        if not self.head: self.head=Node(val); return
        cur=self.head
        while cur.next: cur=cur.next

        cur=self.head
        while cur: print(cur.val,end=' -> '); cur=cur.next

</details>

````python
class Node:
    def __init__(self,val):
      self.val=val
      self.next=None

class LinkedList:
  def __init__(self):
    self.head=None
  def append(self,val):
    if not self.head:
      self.head=Node(val)
      return
    cur=self.head
    while cur.next:
      cur=cur.next
    cur.next=Node(val)
  def print_all(self):
    cur=self.head
    while cur:
      print(cur.val,end=' -> ')
      cur=cur.next
    print('None')
ll=LinkedList()
for v in [1,2,3,4,5]:
  ll.append(v)
ll.print_all()
````

### 419 클래스로 구현한 이진 탐색 트리

이진 탐색 트리의 insert와 inorder traversal을 구현하세요.

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
class TreeNode:
    def __init__(self,val): self.val=val; self.left=self.right=None

class BST:
    def __init__(self): self.root=None
    def insert(self,val):
        def _ins(node,val):
            if not node: return TreeNode(val)
            if val<node.val: node.left=_ins(node.left,val)
            elif val>node.val: node.right=_ins(node.right,val)
            return node
        self.root=_ins(self.root,val)
    def inorder(self):
        result=[]
        def _inorder(node):
            if node: _inorder(node.left); result.append(node.val); _inorder(node.right)
        _inorder(self.root)
        return result

bst=BST()
for v in [5,3,7,1,4,6,8]: bst.insert(v)
print(bst.inorder())  # [1,3,4,5,6,7,8]
</pre>
</details>

````python
class TreeNode:
  def __init__(self,val):
    self.val=val
    self.left=self.right=None

class BST:
  def __init__(self):
    self.root=None
  def insert(self,val):
    def _ins(node,val):
      if not node:
        return TreeNode(val)
      if val<node.val:
        node.left=_ins(node.left,val)
      elif val>node.val:
        node.right=_ins(node.right,val)
      return node
    self.root=_ins(self.root,val)
  def inorder(self):
    result=[]
    def _inorder(node):
      if node:
        _inorder(node.left)
        result.append(node.val)
        _inorder(node.right)
    _inorder(self.root)
    return result

bst=BST()
for v in [5,3,7,1,4,6,8]:
  bst.insert(v)
print(bst.inorder())
````

### 420 Observable 패턴 (이벤트)

Observer 패턴을 구현하세요. 주식 가격이 변할 때 구독자에게 알림을 보내세요.

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
class StockMarket:
    def __init__(self): self._observers=[]; self._price=0
    def subscribe(self,observer): self._observers.append(observer)
    def unsubscribe(self,observer): self._observers.remove(observer)
    def set_price(self,price):
        self._price=price
        for obs in self._observers: obs.update(self._price)

class Investor:
    def __init__(self,name): self.name=name
    def update(self,price): print(f'{self.name}: 현재가 {price:,}원')

market=StockMarket()
kim=Investor('김민수'); lee=Investor('이지영')
market.subscribe(kim); market.subscribe(lee)
market.set_price(70000)
market.unsubscribe(lee)
market.set_price(71500)
</pre>
</details>

````python
class StockMarket:
  def __init__(self):
    self._observers=[]
    self._price=0
  def subscribe(self,observer):
    self._observers.append(observer)
  def unsubscribe(self,observer):
    self._observers.remove(observer)
  def set_price(self,price):
    self._price=price
    for obs in self._observers:
      obs.update(self._price)

class Investor:
  def __init__(self,name):
    self.name=name
  def update(self,price):
    print(f'{self.name}: 현재가 {price:,}원')

market=StockMarket()
kim=Investor('김민수')
lee=Investor('이지영')
market.subscribe(kim)
market.subscribe(lee)
market.set_price(70000)
market.unsubscribe(lee)
market.set_price(71500)
````
