### 294 파일 읽기 - 종목코드 리스트 저장

'매수종목1.txt' 파일을 읽은 후 종목코드를 리스트에 저장해보세요.
```
실행 예:
print(codes)  # ['005930', '005380', '035420']
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
readlines()는 각 줄을 리스트 요소로 반환합니다. strip()으로 줄바꿈(\n)을 제거합니다.</br>
<pre>
f = open("매수종목1.txt", encoding="utf-8")
lines = f.readlines()

codes = []
for line in lines:
    code = line.strip()  # '\n' 제거
    codes.append(code)

print(codes)
f.close()
</pre>
</details>

````python
# 291번에서 매수종목1.txt를 먼저 생성해야 합니다
f = open("매수종목1.txt", encoding="utf-8")
lines = f.readlines() # 파일의 모든 줄을 읽어서 리스트 형태로 저장(예: ['005930\n', '005380\n','035420\n','삼성전자'])

codes=[]

for line in lines: # lines 리스트에 담긴 내용을 한 줄씩 꺼내어 line이라는 변수에 담고 반복
    code = line.strip() # 줄바꿈 문자(\n)나 앞뒤 공백을 깨끗하게 지움 (글자만 남김)
    codes.append(code)

print(codes)
f.close()
````

### 295 파일 읽기 - 딕셔너리로 저장

'매수종목2.txt' 파일을 읽은 후 종목코드와 종목명을 딕셔너리로 저장해보세요. 종목코드를 key, 종목명을 value로 저장합니다.
```
실행 예:
print(data)  # {'005930': '삼성전자', '005380': '현대차', '035420': 'NAVER'}
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
split()으로 공백 기준 분리 후 딕셔너리에 저장합니다.</br>
<pre>
f = open("매수종목2.txt", encoding="utf-8")
lines = f.readlines()

data = {}
for line in lines:
    line = line.strip()  # '\n' 제거
    k, v = line.split()  # '005930 삼성전자' → ['005930', '삼성전자']
    data[k] = v

print(data)
f.close()
</pre>
</details>

````python
# 292번에서 매수종목2.txt를 먼저 생성해야 합니다
f = open("매수종목2.txt", encoding="utf-8")
lines = f.readlines() # 파일의 모든 줄을 읽어서 리스트 형태로 저장(예: ['005930 삼성전자\n', '005380 현대차\n', '035420 NAVER'])

data = {}
for line in lines:
    line = line.strip() # '\n' 제거
    k, v = line.split() ## '005930 삼성전자' → ['005930', '삼성전자']
    data[k] = v # k(코드)를 찾으면 v(이름)가 나오게끔 짝을 맞춰 저장

print(data)
f.close()
print(data['005930'])
````

### 296 예외처리 - try/except 기초

문자열 PER 값을 실수로 변환할 때 에러가 발생합니다. 예외처리를 통해 에러가 발생하는 PER은 0으로 출력하세요.
```python
per = ["10.31", "", "8.00"]
for i in per:
    print(float(i))  # "" → ValueError!
```
```
실행 예 (예외처리 후):
10.31
0
8.0
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
try에 실행 코드, except에 에러 처리 코드를 작성합니다.</br>
<pre>
per = ["10.31", "", "8.00"]

for i in per:
    try:
        print(float(i))
    except:
        print(0)
</pre>
</details>

````python
per = ["10.31", "", "8.00"]

for i in per:
    if i =="":
        print(0)
    else:
        print(float(i)) # "" → ValueError!
print('-'*10)
per = ["10.31", "", "8.00"]

for i in per:
    try:
        print(float(i))
    except:
        print(0)
````

### 298 특정 예외만 처리하기

어떤 값을 0으로 나누면 ZeroDivisionError 에러가 발생합니다. try ~ except로 모든 에러를 처리하지 말고 ZeroDivisionError 에러만 예외처리해보세요.

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
except 뒤에 특정 예외 클래스명을 명시합니다.</br>
<pre>
try:
    b = 3 / 0
except ZeroDivisionError:
    print("0으로 나누면 안되요")
</pre>
</details>

````python
nums = []
for i in range(5):
  try:
    y = 10 / i
  except ZeroDivisionError:
    y = "나눌수 없다!!"
  nums.append(y)
print(nums)
````

### 299 예외 메시지 출력하기

### 300 try / except / else / finally 🎊

파이썬 예외처리의 전체 구조를 사용해보세요.
```python
try:
    실행 코드
except:
    예외가 발생했을 때 수행할 코드
else:
    예외가 발생하지 않았을 때 수행할 코드
finally:
    예외 발생 여부와 상관없이 항상 수행할 코드
```
아래 코드에 예외처리를 적용하고 else, finally에 적당한 메시지를 출력하세요.
```python
per = ["10.31", "", "8.00"]
for i in per:
    print(float(i))  # 오류 발생!
```
```
실행 예:
10.31
clean data
변환 완료
0
변환 완료
8.0
clean data
변환 완료
```

````python
per = ["10.31", "", "8.00"]

# try / except / else / finally 를 모두 사용해보세요!

for i in per:
    try:
      print(float(i))
    except:
      print(0)
    else:
      print('clean data')
    finally:
      print('변환 완료')
````
