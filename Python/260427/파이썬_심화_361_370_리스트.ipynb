### 363 리스트 구간 합

리스트에서 임의 구간 [i:j]의 합을 빠르게 구하는 prefix sum을 구현하세요.
```python
nums=[1,2,3,4,5]
# prefix=[0,1,3,6,10,15]
# range_sum(1,3) = 2+3+4 = 9
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
nums=[1,2,3,4,5]
prefix=[0]
for n in nums: prefix.append(prefix[-1]+n)
def range_sum(i,j): return prefix[j+1]-prefix[i]
print(range_sum(1,3))  # 9
</pre>
</details>

````python
nums=[1,2,3,4,5]
prefix=[0]
for n in nums:
  prefix.append(prefix[-1]+n) #[0, 1, 3, 6, 10, 15]
def range_sum(i,j):
  return prefix[j+1]-prefix[i] # 10-1
print(range_sum(1,3))
````

### 369 리스트 병합 정렬

병합 정렬(merge sort)을 구현하세요.
```python
nums=[5,3,8,1,9,2]
# 결과: [1,2,3,5,8,9]
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
def merge_sort(arr):
    if len(arr)<=1: return arr
    mid=len(arr)//2
    left=merge_sort(arr[:mid])
    right=merge_sort(arr[mid:])
    result=[]
    i=j=0
    while i<len(left) and j<len(right):
        if left[i]<=right[j]: result.append(left[i]); i+=1
        else: result.append(right[j]); j+=1
    return result+left[i:]+right[j:]
nums=[5,3,8,1,9,2]
print(merge_sort(nums))
</pre>
</details>

````python
nums=[5,3,8,1,9,2]
def merge_sort(n):
  if len(n)<=1:
    return n
  mid=len(n)//2
  left=merge_sort(n[:mid])
  right=merge_sort(n[mid:])
  result=[]
  i=j=0
  while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    # 한쪽 리스트가 먼저 끝나면, 남은 다른 쪽 리스트의 요소들을 뒤에 다 붙여줍니다.
  result.extend(left[i:])
  result.extend(right[j:])

  return result

print(merge_sort(nums))
````

### 370 리스트 이상값(outlier) 제거

평균에서 표준편차 2배 이상 벗어난 값을 이상값으로 제거하세요.
```python
data=[10,12,11,13,100,11,12,9,11,10]
# 100은 이상값
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
import math
data=[10,12,11,13,100,11,12,9,11,10]
mean=sum(data)/len(data)
var=sum((x-mean)**2 for x in data)/len(data)
std=math.sqrt(var)
clean=[x for x in data if abs(x-mean)<=2*std]
print('원본:',data)
print('정제:',clean)
</pre>
</details>

````python
import math
data=[10,12,11,13,100,11,12,9,11,10]
mean=sum(data)/len(data)
var=sum((x-mean)**2 for x in data)/len(data)
std=math.sqrt(var)
clean=[x for x in data if abs(x-mean)<=2*std]
print('원본:',data)
print('수정:',clean)
````
