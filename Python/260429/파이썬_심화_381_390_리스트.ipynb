### 381 리스트 원소 연속 여부 확인

리스트가 연속된 정수로 구성되어 있는지 확인하세요.
```python
a=[1,2,3,4,5] → True
b=[1,2,4,5,6] → False
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
def is_consecutive(lst):
    s=sorted(lst)
    return all(s[i+1]-s[i]==1 for i in range(len(s)-1))
print(is_consecutive([1,2,3,4,5]))  # True
print(is_consecutive([1,2,4,5,6]))  # False
</pre>
</details>

````python
a=[1,2,3,4,5]
b=[1,2,4,5,6]

def 연속(lst):
  s=sorted(lst)
  return all(s[i+1]-s[i]==1 for i in range(len(s)-1))
print(연속(a))
print(연속(b))
````

### 382 리스트 원소로 숫자 만들기

숫자 리스트의 원소를 조합해 만들 수 있는 가장 큰 수를 구하세요.
```python
nums=[3,30,34,5,9]
# 결과: 9534330
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
from functools import cmp_to_key
nums=[3,30,34,5,9]
def cmp(a,b): return -1 if str(a)+str(b)>str(b)+str(a) else 1

</details>

````python
from functools import cmp_to_key
nums=[3,30,34,5,9]
def cmp(a,b):
  return -1 if str(a)+str(b)>str(b)+str(a) else 1
print(int(''.join(map(str,sorted(nums,key=cmp_to_key(cmp))))))
````

### 383 리스트 완전 탐색

리스트에서 세 수의 합이 0이 되는 모든 조합을 찾으세요.
```python
nums=[-1,0,1,2,-1,-4]
# 결과: [(-1,-1,2),(-1,0,1)]
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
from itertools import combinations
nums=[-1,0,1,2,-1,-4]
result=set(tuple(sorted(c)) for c in combinations(nums,3) if sum(c)==0)
print(sorted(result))
</pre>
</details>

````python
def three_sum(nums):
    nums.sort()  # 1. 먼저 숫자를 크기순으로 정렬
    results = []

    for i in range(len(nums) - 2):
        # 2. 중복된 첫 번째 숫자는 건너뜀
        if i > 0 and nums[i] == nums[i-1]:
            continue

        # 3. 투 포인터 설정
        left, right = i + 1, len(nums) - 1

        while left < right:
            total = nums[i] + nums[left] + nums[right]

            if total < 0:
                left += 1  # 합이 작으면 왼쪽 포인터를 오른쪽으로
            elif total > 0:
                right -= 1 # 합이 크면 오른쪽 포인터를 왼쪽으로
            else:
                # 합이 0인 경우 결과에 추가
                results.append((nums[i], nums[left], nums[right]))

                # 4. 왼쪽/오른쪽 포인터의 중복 숫자 건너뛰기
                while left < right and nums[left] == nums[left + 1]:
                    left += 1
                while left < right and nums[right] == nums[right - 1]:
                    right -= 1

                left += 1
                right -= 1

    return results

nums = [-1, 0, 1, 2, -1, -4]
print(three_sum(nums)) # 결과: [(-1, -1, 2), (-1, 0, 1)]
````

### 384 리스트 - 주식 이익 계산

주가 리스트가 주어질 때, 한 번의 매수/매도로 얻을 수 있는 최대 이익을 구하세요.
```python
prices=[7,1,5,3,6,4]
# 결과: 5 (1에 사서 6에 팔기)
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
prices=[7,1,5,3,6,4]
min_p=float('inf')
max_profit=0
for p in prices:
    min_p=min(min_p,p)
    max_profit=max(max_profit,p-min_p)
print(max_profit)
</pre>
</details>

````python
prices = [7, 1, 5, 3, 6, 4]
min_p = float('inf')
max_profit = 0

for p in prices:
    # 내장 함수 min, max가 정상 작동하는지 확인
    min_p = min(min_p, p)
    max_profit = max(max_profit, p - min_p)

print(f"최대 이익: {max_profit}")
````

### 385 리스트 - 빗물 저장량 계산

각 기둥의 높이가 주어질 때 빗물이 얼마나 고이는지 계산하세요.
```python
height=[0,1,0,2,1,0,1,3,2,1,2,1]
# 결과: 6
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
height=[0,1,0,2,1,0,1,3,2,1,2,1]
n=len(height)
left_max=[0]*n; right_max=[0]*n
for i in range(1,n): left_max[i]=max(left_max[i-1],height[i-1])
for i in range(n-2,-1,-1): right_max[i]=max(right_max[i+1],height[i+1])
water=sum(max(0,min(left_max[i],right_max[i])-height[i]) for i in range(n))
print(water)
</pre>
</details>

````python
height=[0,1,0,2,1,0,1,3,2,1,2,1]
n=len(height)
left_max=[0]*n; right_max=[0]*n
for i in range(1,n): left_max[i]=max(left_max[i-1],height[i-1])
for i in range(n-2,-1,-1): right_max[i]=max(right_max[i+1],height[i+1])
water=sum(max(0,min(left_max[i],right_max[i])-height[i]) for i in range(n))
print(water)
````

### 386 리스트 - 가장 긴 증가 부분 수열 길이

리스트에서 가장 긴 증가하는 부분 수열(LIS)의 길이를 구하세요.
```python
nums=[10,9,2,5,3,7,101,18]
# 결과: 4 (2,3,7,18 또는 2,5,7,18)
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
nums=[10,9,2,5,3,7,101,18]
import bisect
tails=[]
for n in nums:
    pos=bisect.bisect_left(tails,n)
    if pos==len(tails): tails.append(n)
    else: tails[pos]=n
print(len(tails))
</pre>
</details>

````python
nums=[10,9,2,5,3,7,101,18]
import bisect
tails=[]
for n in nums:
    pos=bisect.bisect_left(tails,n)
    if pos==len(tails): tails.append(n)
    else: tails[pos]=n
print(len(tails))
````

### 388 리스트 - 연속 구간 찾기

nums=[1,1,1,2,2,3,3,3,3]
ranges=[]

### 389 리스트 - 문자열 아나그램 그룹화

단어 리스트에서 아나그램끼리 그룹화하세요.
```python
words=['eat','tea','tan','ate','nat','bat']
# 결과: [['eat','tea','ate'],['tan','nat'],['bat']]
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
from collections import defaultdict
words=['eat','tea','tan','ate','nat','bat']
groups=defaultdict(list)
for w in words:
    groups[tuple(sorted(w))].append(w)
print(list(groups.values()))
</pre>
</details>

````python
words=['eat','tea','tan','ate','nat','bat']
from collections import defaultdict
groups=defaultdict(list)
for w in words:
    groups[tuple(sorted(w))].append(w)
print(list(groups.values()))
````

### 390 리스트 - 부분집합 생성

주어진 리스트의 모든 부분집합을 출력하세요.
```python
nums=[1,2,3]
# [[],[1],[2],[3],[1,2],[1,3],[2,3],[1,2,3]]
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
from itertools import combinations
nums=[1,2,3]
result=[]
for r in range(len(nums)+1):
    result.extend(combinations(nums,r))
print([list(s) for s in result])
</pre>
</details>

````python
from itertools import combinations
nums=[1,2,3]
result=[]
for r in range(len(nums)+1):
    result.extend(combinations(nums,r))
print([list(s) for s in result])
````
