### 351 리스트 이진 탐색

정렬된 리스트에서 bisect 모듈로 값의 삽입 위치를 찾으세요.
```python
sorted_nums=[1,3,5,7,9]
bisect(sorted_nums, 4) → 2
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
import bisect
sorted_nums=[1,3,5,7,9]
print(bisect.bisect_left(sorted_nums, 4))  # 2
print(bisect.bisect_right(sorted_nums, 5)) # 3
</pre>
</details>

````python
import bisect
sorted_nums=[1,3,5,7,9]
print(bisect.bisect_left(sorted_nums, 4)) # 4가 정렬 상태를 유지하며 들어갈 수 있는 가장 왼쪽 인덱스인 2(숫자 5의 자리)를 반환
print(bisect.bisect_right(sorted_nums, 5)) # 5가 정렬 상태를 유지하며 들어갈 수 있는 가장 오른쪽 인덱스인 3(숫자 7의 자리)을 반환
````

### 357 리스트 교집합·합집합·차집합

```python
a=[1,2,3,4,5]; b=[3,4,5,6,7]
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
a=[1,2,3,4,5]; b=[3,4,5,6,7]
b_set=set(b)
inter=[x for x in a if x in b_set]
union=a+[x for x in b if x not in set(a)]

### 358 리스트를 이용한 스택 구현

리스트로 스택을 구현하고 push, pop, peek, is_empty를 사용해보세요.

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
class Stack:
    def __init__(self): self.data=[]
    def push(self,v): self.data.append(v)
    def pop(self): return self.data.pop() if self.data else None
    def peek(self): return self.data[-1] if self.data else None
    def is_empty(self): return len(self.data)==0

s=Stack()
s.push(1); s.push(2); s.push(3)
print(s.peek())   # 3
print(s.pop())    # 3
print(s.is_empty()) # False
</pre>
</details>

````python
class Stack:
  def __init__(self):
    self.data=[]
  def push(self, d):
    self.data.append(d)
  def pop(self):
    return self.data.pop() if self.data else None
  def peek(self):
    return self.data[-1] if self.data else None
  def is_empty(self):
    return len(self.data)==0

s=Stack()
s.push(1)
s.push(2)
s.push(3)
print(s.peek())   # 3
print(s.pop())    # 3
print(s.is_empty()) # False
````

### 359 리스트를 이용한 큐 구현

collections.deque를 사용해 큐(FIFO)를 구현하세요.

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
from collections import deque
class Queue:
    def __init__(self): self.data=deque()
    def enqueue(self,v): self.data.append(v)
    def dequeue(self): return self.data.popleft() if self.data else None
    def peek(self): return self.data[0] if self.data else None

q=Queue()
q.enqueue('A'); q.enqueue('B'); q.enqueue('C')
print(q.dequeue())  # A
print(q.peek())     # B
</pre>
</details>

````python
from collections import deque
class Queue:
    def __init__(self): self.data=deque()
    def enqueue(self,v): self.data.append(v)
    def dequeue(self): return self.data.popleft() if self.data else None
    def peek(self): return self.data[0] if self.data else None

q=Queue()
q.enqueue('A'); q.enqueue('B'); q.enqueue('C')
print(q.dequeue())  # A
print(q.peek())     # B
````

### 360 2차원 리스트 전치

2차원 리스트(행렬)를 전치(transpose)하세요.
```python
matrix=[[1,2,3],[4,5,6]]
# 결과: [[1,4],[2,5],[3,6]]
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
matrix=[[1,2,3],[4,5,6]]

</details>

````python
matrix=[[1,2,3],[4,5,6]]
transposed=list(map(list,zip(*matrix)))
print(transposed)  # [[1,4],[2,5],[3,6]]
````
